```python
import pandas as pd
import numpy as np
import xgboost as xgb
import matplotlib.pyplot as plt
```

```python
train = pd.read_csv('../moblie_data/train_clean.csv', )
test = pd.read_csv('../moblie_data/test_clean.csv')
```

```python
train = train.iloc[:,1:]
test = test.iloc[:,1:]
```

```python
train['current_service'] = train['current_service'].astype(str)
train['gender'] = train['gender'].astype('str')
test['gender'] = test['gender'].astype(str)
train[['contract_type','net_service']] = train[['contract_type','net_service']].astype(str)
test[['contract_type','net_service']] = test[['contract_type','net_service']].astype(str)
```

```python
train.head()
```

```python
xgb_train = pd.get_dummies(train.iloc[:,:-2],drop_first=True)
xgb_label = train.current_service
```

```python
xgb_train = pd.concat([xgb_train,xgb_label],axis=1)
```

```python
train_1 = xgb_train[train.service_type == 1].iloc[:,1:]
train_3 = xgb_train[train.service_type == 3].iloc[:,1:]
train_4 = xgb_train[train.service_type == 4].iloc[:,1:]

test_1 = test[test.service_type == 1].iloc[:,1:]
test_3 = test[test.service_type == 3].iloc[:,1:]
test_4 = test[test.service_type == 4].iloc[:,1:]
```

```python
test_1['predict'] = '0'
test_3['predict'] = '99104722'
test_4['predict'] = '0'
```

```python
train_1.head()
```

## xgboost建模(type1 and type4)

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(train_1.iloc[:,:-1], 
                                                    train_1.iloc[:,-1].astype('str'), random_state=0)

X_total = train_1.iloc[:,:-1]
y_total = train_1.iloc[:,-1]
```

```python
y_train.value_counts()
```

```python
## needs code viewed
y_train[y_train == '90063345'] = 0
y_train[y_train == '90109916'] = 1
y_train[y_train == '90155946'] = 2

y_test[y_test == '90063345'] = 0
y_test[y_test == '90109916'] = 1
y_test[y_test == '90155946'] = 2

y_total[y_total == '90063345'] = 0
y_total[y_total == '90109916'] = 1
y_total[y_total == '90155946'] = 2
```

```python
y_train.shape
```

```python
y_train.value_counts()
```

```python
X_train.shape
```

```python
dtrain = xgb.DMatrix(X_train,label=y_train)
dtest = xgb.DMatrix(X_test,label=y_test)

dtotal = xgb.DMatrix(X_total,label=y_total)
```

```python
param = {'max_depth':5 , 'silent':1, 'objective':'multi:softprob','subsample':0.9, 'colsample_bytree':1,
         'learning_rate':0.1, 'nthread':4, 'min_child_weight':2 ,
         'lambda':0, 'alpha':1, 'gamma':0,'num_class':3}
#param['eval_metric'] = 'auc'
evallist = [(dtrain,'train'),(dtest,'eval')]
```

```python
num_round = 500
bst = xgb.train(param.items(), dtrain, num_round, evals = evallist, early_stopping_rounds = 30)
```

```python
#全部数据训练
num_round = 200
bst = xgb.train(param.items(), dtotal, num_round)
```
