```python
import pandas as pd
import numpy as np
import xgboost as xgb
import matplotlib.pyplot as plt
```

```python
train = pd.read_csv('../moblie_data/train_clean.csv', )
test = pd.read_csv('../moblie_data/test_clean.csv')
```

```python
train = train.iloc[:,1:]
test = test.iloc[:,1:]
```

```python
train['current_service'] = train['current_service'].astype(str)
train['gender'] = train['gender'].astype('str')
test['gender'] = test['gender'].astype(str)
train[['contract_type','net_service']] = train[['contract_type','net_service']].astype(str)
test[['contract_type','net_service']] = test[['contract_type','net_service']].astype(str)
```

```python
train.head()
```

```python
xgb_train = pd.get_dummies(train.iloc[:,:-2],drop_first=True)
xgb_label = train.current_service
```

```python
xgb_test = pd.get_dummies(test.iloc[:,:-1],drop_first=True)
```

```python
xgb_train = pd.concat([xgb_train,xgb_label],axis=1)
```

```python
train_1 = xgb_train[train.service_type == 1].iloc[:,1:]
train_3 = xgb_train[train.service_type == 3].iloc[:,1:]
train_4 = xgb_train[train.service_type == 4].iloc[:,1:]

test_1 = xgb_test[test.service_type == 1].iloc[:,1:]
test_3 = xgb_test[test.service_type == 3].iloc[:,1:]
test_4 = xgb_test[test.service_type == 4].iloc[:,1:]
```

## xgboost建模(type1 and type4)

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(train_4.iloc[:,:-1], 
                                                    train_4.iloc[:,-1].astype('str'), random_state=0)

X_total = train_4.iloc[:,:-1]
y_total = train_4.iloc[:,-1]
```

```python
y_train.value_counts()
```

```python
## needs code viewed
def change_class_1(y):
    y[y == '90063345'] = 0
    y[y == '90109916'] = 1
    y[y == '90155946'] = 2
    return y
    
def change_class_4(y):
    y[y == '89950166'] = 0
    y[y == '89950167'] = 1
    y[y == '99999828'] = 2
    y[y == '89016252'] = 3
    y[y == '89950168'] = 4
    y[y == '99999827'] = 5
    y[y == '99999826'] = 6
    y[y == '99999830'] = 7
    y[y == '99999825'] = 8
    y[y == '89016253'] = 9
    y[y == '89016259'] = 10
    return y

y_train = change_class_4(y_train)
y_test = change_class_4(y_test)
y_total = change_class_4(y_total)
```

```python
y_train.shape
```

```python
y_train.value_counts()
```

```python
X_train.shape
```

```python
dtrain = xgb.DMatrix(X_train,label=y_train)
dtest = xgb.DMatrix(X_test,label=y_test)

dtotal = xgb.DMatrix(X_total,label=y_total)
```

```python
param = {'max_depth':5 , 'silent':1, 'objective':'multi:softprob','subsample':0.9, 'colsample_bytree':1,
         'learning_rate':0.3, 'nthread':4, 'min_child_weight':2 ,
         'lambda':0, 'alpha':1, 'gamma':0,'num_class':11}
#param['eval_metric'] = 'auc'
evallist = [(dtrain,'train'),(dtest,'eval')]
```

```python
%%time
num_round = 500
bst = xgb.train(param.items(), dtrain, num_round, evals = evallist, early_stopping_rounds = 30)
```

```python
#全部数据训练
num_round = 400
bst = xgb.train(param.items(), dtotal, num_round)
```

```python
bst.save_model('xgb_type4.model')
```

## 预测，整理表格

```python
pred_4 = bst.predict(xgb.DMatrix(test_4))
```

```python
def submit(pred, type):
    df_p1 = pd.DataFrame(pred)
    p1_idmax = df_p1.idxmax(axis=1)
    def change_class_1(y):
        y[y == 0] = 90063345
        y[y == 1] = 90109916
        y[y == 2] = 90155946
        return y

    def change_class_4(y):
        y[y == 0] = 89950166
        y[y == 1] = 89950167
        y[y == 2] = 99999828
        y[y == 3] = 89016252
        y[y == 4] = 89950168
        y[y == 5] = 99999827
        y[y == 6] = 99999826
        y[y == 7] = 99999830
        y[y == 8] = 99999825
        y[y == 9] = 89016253
        y[y == 10] = 89016259
        return y

    p1_idmax = change_class_4(p1_idmax)
    
    final_pred = test[test.service_type == type][['user_id']]
    final_pred.reset_index(drop=True, inplace=True)
    final_pred['predict'] = p1_idmax
    
    return final_pred

final_pred_4 = submit(pred_4, 4)
```
